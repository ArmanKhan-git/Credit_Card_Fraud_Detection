```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

# Info about data

```python
df.info()
```

```python
df.describe()
```

```python
df=df.dropna(subset=['V23','V24','V25','V26','V27','V28','Amount','Class'])
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

# Data is highly imbalance

```python
# seperating the data into legit and fraud
legit=df[df['Class']==0]
fraud=df[df['Class']==1]
```

```python
# now statistical measures of the data
print(legit['Amount'].describe())
print(fraud['Amount'].describe())
```

```python
df.groupby('Class').mean()
```

```python
# scaling amount and time
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
cols=['Time','Amount']
df[cols]=scaler.fit_transform(df[cols])
```

# Now due to imbalance we will take a sample dataset but first we gotta split the dataset cause if i done balancing now then it wont reflect to the real world scenario so i gotta do it in train dataset only


```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['Class'])
y=df['Class']
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
y_res.value_counts()
```

# Taking out the sample from this population

```python
from sklearn.model_selection import train_test_split

# Let's take 50k rows as an example
X_sample, _, y_sample, _ = train_test_split(
    X_res, y_res,
    train_size=50000,       # total rows in sample
    stratify=y_res,         # maintain same class ratio
    random_state=42
)

```

```python
y_sample.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.20,stratify=y_sample, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score,precision_score,recall_score,roc_auc_score,confusion_matrix
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB()
}

results=[]
```

```python
for name,model in models.items():
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  results.append({
      'model':name,
      'accuracy': accuracy_score(y_test, y_pred),
      'f1': f1_score(y_test, y_pred),
      'Precision':precision_score(y_test,y_pred),
      'Recall':recall_score(y_test,y_pred),
      'ROC-AUC':roc_auc_score(y_test,y_pred)
  })
```

```python
results
```

```python
results_train=[]
for name,model in models.items():
  results_train.append({
      'model':name,
      'score': model.score(X_train,y_train)
  })
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier_knn = GridSearchCV(
    KNeighborsClassifier(),
    param_grid={
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['minkowski', 'euclidean', 'manhattan']
    },
    cv=5,
    return_train_score=False
)
```

```python
classifier_knn.fit(X_train,y_train)
```

```python
y_pred_best = classifier_knn.best_estimator_.predict(X_test)
y_prob_best = classifier_knn.best_estimator_.predict_proba(X_test)[:, 1]
```

```python
acc=accuracy_score(y_test,y_pred_best)
f1=f1_score(y_test,y_pred_best)
recall=recall_score(y_test,y_pred_best)
precision=precision_score(y_test,y_pred_best)
Roc_score=roc_auc_score(y_test,y_prob_best)
```

```python
print(acc)
print(f1)
print(recall)
print(precision)
print(Roc_score)
```
